```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('hasil_pivot2019.xlsx')
```

```python
df.head(3)
```

```python
df.isna().sum()
```

```python
df=df[['KECAMATAN','CABUL','CURANMOR','CURAS','CURAT','KDRT','PELECEHAN SEKSUAL','PEMALSUAN','PEMBAKARAN',
        'PEMBUNUHAN','PEMERKOSAAN','PENCURIAN','PENCURIAN DAN PENGGELAPAN','PENGANIAYAAN','PENGEROYOKAN','PENGERUSAKAN',
       'PENGGELAPAN','PENIPUAN','PENIPUAN DAN PENGGELAPAN','PERUNDUNGAN ANAK']]
```

```python
df.head(2)
```

```python
print(df.shape)
```

```python
# Cek apakah ada yang kosong datanya
df.isna().sum()
# jika ada yang kosong, maka bersihkan jika tidak maka comment code d bawah
# df.dropna(inplace=True)
dt_processing = df.drop('KECAMATAN', axis=1)
dt_processing
```

```python
# menentukan nilai minimum untuk centroid 1 awal
index_min = dt_processing.idxmin()['CABUL']
c1 = dt_processing.iloc[[index_min]].values
c1
```

```python
# menentukan nilai rata-rata untuk centroid 2 awal
index_average = dt_processing.mean()['CABUL']
c2 = dt_processing.iloc[[index_average]].values
c2
```

```python
# menentukan nilai maximum untuk centroid 3 awal
index_max = dt_processing.idxmax()['CABUL']
c3 = dt_processing.iloc[[index_max]].values
c3
```

```python
# gabungkan c1 , c2 dan c3
nilai_centroid_awal = np.concatenate((c1, c2, c3), axis=0)
nilai_centroid_awal
```

```python
# mulai proses kmeans nya
X = np.array(nilai_centroid_awal, np.float64)
km = KMeans(n_clusters=3, init=X, n_init=1).fit(dt_processing)
centers = km.cluster_centers_
print(centers)
```

```python
# masukan nilai cluster ke dalam dataframe
df['cluster'] = km.labels_
# Code
a = km.labels_
mapping = {0:'Tidak Rawan', 1:'Cukup Rawan', 2:'Rawan'}
a = [mapping[i] for i in a]
df['keterangan'] = a
```
